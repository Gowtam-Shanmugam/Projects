```python
# extracting out the columns with object dataformat.

col_list=[]  # empty list

for i in churn_data.columns:

  if (churn_data[i].dtype=='object'):
    col_list.append(i)
```

```python
# Label Encoding:

from sklearn.preprocessing import LabelEncoder
```

```python
sky=LabelEncoder()
```

```python
for j in col_list:
  churn_data[j]=sky.fit_transform(churn_data[j])
```

```python
churn_data['Churn']=churn_data['Churn'].map({0:'No',1:'Yes'})
```

```python
# Model Building:

# forming up the dependent and independent variable.
# Spliting criteria with training and testing buckets.


x= churn_data.iloc[:,:-1]   # R----> All rows, c---> All columns till second last.

y=churn_data.iloc[:,-1]   # R---> All rows, c---> -1 (Last column)
```

```python
x
```

```python
y
```

```python
# Following 80---> Training  Bucket
#           20---> Testing Bucket.

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

```

```python
x_train.head()
```

```python
# prepare the model.
# Hyperparameter : Parameter controlled the developer or the programmer--> max_dept, estimato
from sklearn.ensemble import RandomForestClassifier
```

```python
Model1=RandomForestClassifier(n_estimators=500)
```

```python
Model1.fit(x_train,y_train)
```

```python
x_test.head()
```

```python
y_pred=Model1.predict(x_test)
```

```python
# Hyperparameter 

n=[100,200,300,400,500,600,700,800,900,1000]
```

```python
thunder=[]  # storing the accurancy

for h in n:
  model=RandomForestClassifier(n_estimators=h)
  model.fit(x_train,y_train)
  pred=model.predict(x_test)
  a=accuracy_score(y_test,pred)
  thunder.append(a)
```

```python
choice={'Number of tree':n,'Accuracy':thunder}
```
