```python
# PCA is not a machine learning technique, its a unsupervised machine learning algorithm which will help you to reduce the number of dimensions 
# or the number of features into lower number of dimensions 
# with the increase number in dimension accuracy will be impacted and less
# Converting 2-D to 1-D
# principal component 1 = best fit line, principal component 2 and so on
# variance lost should be very less
```

```python
# First Step in PCA - Standard Scaler or Standard normal distribution where mean will be zero and standard deviation will be 1
# rescaling all the values in the same range 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
```

```python
scaled_data
```

```python
scaled_data=scaler.transform(df)
```

```python
print(scaled_data)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
x_pca=pca.fit_transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer["target"],cmap='plasma')
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler  # standardization
scaler=StandardScaler()
scaled_data=scaler.fit_transform(X)
```

```python
scaled_data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(scaled_data,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=29)
x_pca=pca.fit_transform(scaled_data)
```

```python
x_pca.shape
```

```python
x_pca
```

```python
from sklearn.model_selection import train_test_split
x_train_pca,x_test_pca,y_train,y_test= train_test_split(x_pca,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(x_train_pca,y_train)
model.score(x_test_pca,y_test)
```
